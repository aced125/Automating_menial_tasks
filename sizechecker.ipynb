```python
# Make a list of only the top-level sub-directories of a directory

folder_list = [name for name in os.listdir(local_path)]
```

```python
def get_size(path):
    
    # Loops through all subdirectories within
    # a directory and sums all their sizes.
    
    # The try and except statements are in place
    # because sometimes Windows doesn't allow
    # certain operations on certain files
    
    total_size = 0
    for root, dirs, files in os.walk(path, topdown = False):
        for name in files:
            try:
                fp = os.path.join(root,name)
                total_size += os.path.getsize(fp)
            except:
                pass
    return total_size
```

```python
# List comprehension to get sizes of all the top-level subdirectories of a directory
size_list = [get_size(os.path.join(local_path, folder_list[i])) 
             for i in range(len(folder_list))]

# Convert to numpy array
size_list = np.array(size_list)
```

```python
# Sort the list by size (the [::-1] reverses the order of the list,
# giving descending order)

sorted_size_list = np.sort(size_list)[::-1]
sorted_idx_list = np.argsort(size_list)[::-1]
```

```python
# Create a list of sorted folder names with list comprehension

sorted_folder_names = [folder_list[i] for i in sorted_idx_list]
```

```python
# Create a Pandas dataframe.

# Pass in the data as a list of lists.
# Transpose dataframe to get into correct format
# (ie folder names as a column rather than row).

# Set the column names appropriately


df = pd.DataFrame(
    data = [sorted_folder_names,
            sorted_size_list/10**9]).T

df.columns = ['Folder','Size / Gb']
```

```python
def find_problem_folders(local_path):
    
    folder_list = [name for name in os.listdir(local_path) 
               if os.path.isdir(os.path.join(local_path, name)) ]
    
    
    # List comprehension to get sizes of all the top-level subdirectories of a directory
    size_list = [get_size(os.path.join(local_path, folder_list[i])) 
                 for i in range(len(folder_list))]

    # Convert to numpy array
    size_list = np.array(size_list)
    
    # Sort the list by size (the [::-1] reverses the order of the list,
    # giving descending order)
    sorted_size_list = np.sort(size_list)[::-1]
    sorted_idx_list = np.argsort(size_list)[::-1]
    
    # Create a list of sorted folder names with list comprehension
    sorted_folder_names = [folder_list[i] for i in sorted_idx_list]
    
    # Create a Pandas dataframe.

    # Pass in the data as a list of lists.
    # Transpose dataframe to get into correct format
    # (ie folder names as a column rather than row).

    # Set the column names appropriately


    df = pd.DataFrame(
        data = [sorted_folder_names,
                sorted_size_list/10**9]).T

    df.columns = ['Folder/File','Size / Gb']
    
    return df
```
